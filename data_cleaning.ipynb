# Combining News Articles by Month

```python
nonempty_consumer_news_data = pd.read_csv('data/news/nonempty_consumer_news_data.csv')
```

```python
import datetime as dt
nonempty_consumer_news_data['date'] = [dt.datetime.strptime(time, '%Y-%m-%d %H:%M:%S.%f') for time in nonempty_consumer_news_data['TIMESTAMP_UTC']]
nonempty_consumer_news_data['date']
```

```python
grouped_news = nonempty_consumer_news_data.groupby([pd.Grouper(key='date',freq='M'), 'RP_ENTITY_ID']).mean().reset_index()
grouped_news.head()
```

```python
ticker_mappings = pd.read_csv('data/rp_full_ticker_mapping/ticker_mappings.csv')
ticker_mappings.head()
```

```python
ticker_mappings.loc[ticker_mappings['security'] == 'NKE US Equity']['RP_ENTITY_ID'][0]
```

```python
grouped_news['security'] = [list(ticker_mappings.loc[ticker_mappings['RP_ENTITY_ID'] == grouped_news['RP_ENTITY_ID'][i]]['security'])[0] for i in range(len(grouped_news['RP_ENTITY_ID']))]
```

```python
grouped_news
```

```python
financials = pd.read_csv('data/financials.csv')
financials
```

```python
def date_to_month(date):
    if type(date) == str:
        return '-'.join(date.split('-')[:2])
    return str(date.year) + '-' + str(date.month)
financials['month'] = [date_to_month(date) for date in financials['date']]
grouped_news['month'] = [date_to_month(date) for date in grouped_news['date']]
```

```python
news_and_finance_data = pd.merge(financials, grouped_news, how='left', left_on=['month','security'], right_on = ['month','security'])
news_and_finance_data
```

```python
news_and_finance_data.fillna(0)
```

```python
grouped_news.to_csv('data/news/grouped_news.csv', index=False)
news_and_finance_data.to_csv('data/news_and_finance_date.csv', index=False)
```
